```python
#ratings
ratings=[]

rating=driver.find_elements_by_xpath("//div[@class='rating rating-summary-container big']")
for i in rating:
    if i.text is None :
        ratings.append("--") 
    else:
        ratings.append(i.text[0:3])
ratings
```
